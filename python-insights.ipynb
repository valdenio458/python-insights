```python
import plotly.express as px
```

```python
for coluna in tabela1.columns:
    grafico = px.histogram(tabela1, x=coluna, color="cancelou", text_auto=True)
    grafico.show()
```

```python
# Usuários com contrato mensal sempre cancelam
  # Evitar o contrato mensal e incentivar, com decontos, os outros tipos de contrato

# Todos os usários que ligaram mais de 4 vezes para o call center cancelaram o serviço

# Usuários que atrasaram o pagamento mais de 20 dias cancelaram

# Soluções:
# duracão_contrato -> diferente de mensal
# ligacoes_callcenter -> menor que 3
# atraso_pagamento -> menor que 20 dias
```
